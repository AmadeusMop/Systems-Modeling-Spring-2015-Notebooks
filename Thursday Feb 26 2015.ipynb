## SIRS Model

```python
sim=Simulation()
sim.add("S'=-a*S*I+c*R", 100, plot=1)
sim.add("I'=+a*S*I-b*I", 1, plot=1)
sim.add("R'=+b*I-c*R", 0, plot=1)
sim.params(a=.1,b=1,c=1)
sim.run(0,10)
```
