### Linear growth example

$y' = {\rm constant}$

```python
y=2
t=0

constant=10
dt=0.01
```

```python
reset()

store(t,y)
for i in range(9000):
    dy=constant*dt
    
    y=y+dy
    t=t+dt
    
    if i%100 == 0 :  # if the remainder with i/100 = 0 (i.e. i is divisible by 100)
        store(t,y)
```

```python
t,y=recall()
```

```python
plot(t,y)
xlabel('time')
ylabel('y')
```

## Answer to the HW question

```python
x=2
v=1
t=0

k=10
c=0.1
dt=0.0005  # I made the time step smaller, so I can see the beginning
```

```python
reset()

store(t,x,v)
for i in range(9000):
    dx=v*dt
    dv=(k-c*v**2)*dt
    
    x=x+dx
    v=v+dv
    t=t+dt
    
    if i%100 == 0 :  # if the remainder with i/100 = 0 (i.e. i is divisible by 100)
        store(t,x,v)
```

```python
t,x,v=recall()
```

```python
plot(t,x)
xlabel('time')
ylabel('x')
```

```python
plot(t,v)
xlabel('time')
ylabel('v')
```
