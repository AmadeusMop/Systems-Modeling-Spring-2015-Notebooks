```python
from science import *
```

```python
x=linspace(0,20,100)
y=0.1*x+3+randn(len(x))
plot(x,y,'o')
```

```python
def constant(x,a):
    return a
```

```python
model=MCMCModel(x,y,constant,
                a=Uniform(-10,10))
model.run_mcmc(500)
model.plot_chains()
```

```python
model.plot_distributions()
```

## Linear Model

```python
def linear(x,m,b):
    return m*x+b
```

```python
model=MCMCModel(x,y,linear,
                m=Uniform(-10,10),
                b=Uniform(-10,10),
                )
model.run_mcmc(500)
model.plot_chains()
```

```python
model.plot_distributions()
```

## What happens if the true value is outside the prior?

```python
x=linspace(0,20,100)
y=5*x+30+4*randn(len(x))
plot(x,y,'o')
```

```python
model=MCMCModel(x,y,linear,
                m=Uniform(-10,10),
                b=Uniform(-10,10),
                )
model.run_mcmc(500)
model.plot_chains()
```

modify the range of the prior...

```python
model=MCMCModel(x,y,linear,
                m=Uniform(-10,10),
                b=Uniform(-100,100),   # <---- bigger range
                )
model.run_mcmc(500)
model.plot_chains()
```

```python
model.plot_distributions()
```

```python
model.triangle_plot()
```
