```python
from pyndamics import *
from pyndamics.emcee import *
```

```python
x=linspace(0,20,100)
y=0.1*x+3+randn(len(x))
plot(x,y,'o')
```

```python
sim=Simulation()
sim.add("y'=a",1,plot=True)
sim.params(a=0.05)
sim.add_data(t=x,y=y,plot=True)
sim.run(0,20)
```

```python
model=MCMCModel(sim,a=Uniform(-1,1))
```

```python
model.run_mcmc(500)
model.plot_chains()
```
