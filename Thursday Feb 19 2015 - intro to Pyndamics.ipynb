## yet another example - logistic!

* $y'=ay\cdot (K-y)$
* $K=100$
* $a=0.02$
* $t$ from 0 to 90
* initia value of $y=0.1$

```python
sim=Simulation()
sim.add("y'=a*y*(1-y/K)",0.1,plot=True)
sim.params(a=.02,K=100)
sim.run(0,90)
```

what happens when I run longer?

```python
sim=Simulation()
sim.add("y'=a*y*(1-y/K)",0.1,plot=True)
sim.params(a=.02,K=100)
sim.run(0,900)
```
